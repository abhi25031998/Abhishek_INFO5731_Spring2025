# Question 1 (10 points).

The formula that describes the volume V of a hemisphere with radius r is the following:

$ V=\frac{2}{3}\ast\ \pi\ast\ r^3 $

Write a program to calculate the value of V when r is in the range of even numbers from 1-20. Output the result in the following format:

	r	V
	2	…
	4	…
	6	…
	.
	.
	.
	20	…


```python
import math
#Prints the header
print("r        V") #can also use '\t' instead of space

for r in range(2, 21, 2): #iterates alternate number(even) btw 2-20
    V = (2 / 3) * math.pi * r ** 3 # calculating volume of Hemisphere
    #insert values directly into string and V is set to two decimal values
    print(f"{r}\t{V:.2f}")
```

```python
# Write code for each of the sub parts with proper comments.
import random
import numpy as np

#Generate and print 12 random numbers between 50 and 150.
print("Sequence of 12 random numbers between 50 and 150: ")
rand_seq_list = []
for i in range(12):
     rand_seq = random.randrange(50,150)
     rand_seq_list.append(rand_seq)
     print(rand_seq)

#arrange the numbers in ascending order.
rand_seq_list.sort()
print("Ascending order of the sequence:", rand_seq_list)

#average of first six numbers of the ascending order.
avg_first6_asc = np.mean(rand_seq_list[:6])

#arrange the numbers in descending order.
rand_seq_list.sort(reverse=True)
print("Descending order of the sequence:", rand_seq_list)

#finding the average of the all 12 generated numbers.
avg = np.mean(rand_seq_list)
print("Average of the sequence:" , f"{avg:.3f}")

#average of first six numbers in descending order.
avg_first6_desc = np.mean(rand_seq_list[:6])

#display the calculated average of the first 6 numbers in ascending and descending order.
print("Average of the first 6 numbers in ascending order:", f"{avg_first6_asc:.3f}")
print("Average of the first 6 numbers in descending order:", f"{avg_first6_desc:.3f}")


```

# Question 4 (10 Points)

Design and implement a simple command-line calculator program. The program should handle both integers and floating-point numbers.

(a) Write program that takes a mathematical expression as input and evaluates it. The program should support addition, subtraction, multiplication, and division.

(b) Extend the calculator to handle parentheses in mathematical expressions. Implement the necessary logic to evaluate expressions with nested parentheses.

(c) Enhance your calculator to include additional mathematical operations such as exponentiation, square root, and modulus.

(d) Implement error handling in the calculator program. Handle cases such as division by zero and invalid input expressions. Provide meaningful error messages.

(e) Implement a command to exit the calculator.

```python
import math
#compute or evaluate the given equations
def evaluate_expression(expression):
    try:
        safe_dict = {
            "__builtins__": None,
            "sqrt": math.sqrt,
            "pow": math.pow,
            "abs": abs,
            "round": round,
            "math": math
        }
        result = eval(expression, safe_dict)
        return result

    except Exception as e:
        return f"Error: {e}"

#choose and caculate the operation to perform
def calculate():
    operation = input('''
Please type in the math operation you would like to complete:
+ for addition
- for subtraction
* for multiplication
/ for division
** for exponentiation
% for modulus
sqrt: for square root
expr: for evaluating an expression
exit: to exit the calculator
''')

    if operation == 'exit':
        print('Exiting calculator. See you later, Goodbye')
        return

    if operation == 'sqrt':
        number = float(input('Please enter a number: '))
        print('sqrt({}) = {}'.format(number, math.sqrt(number)))
        again()

    elif operation == 'expr':
        expression = input("Enter a mathematical expression: ")
        result = evaluate_expression(expression)
        print(f"Result: {result}")
        again()

    else:
        number_1 = float(input('Please enter the first number: '))
        number_2 = float(input('Please enter the second number: '))

        if operation == '+':
            print('{} + {} = '.format(number_1, number_2))
            print(number_1 + number_2)
        elif operation == '-':
            print('{} - {} = '.format(number_1, number_2))
            print(number_1 - number_2)
        elif operation == '*':
            print('{} * {} = '.format(number_1, number_2))
            print(number_1 * number_2)
        elif operation == '**':
            print('{} ** {} = '.format(number_1, number_2))
            print(number_1 ** number_2)
        elif operation == '%':
            print('{} % {} = '.format(number_1, number_2))
            print(number_1 % number_2)
        elif operation == '/':
            if number_2 == 0:
                print("Error: Division by zero is not allowed.")
            else:
                print('{} / {} = '.format(number_1, number_2))
                print(number_1 / number_2)
        else:
            print('Invalid operator, please choose a valid operator.')
        again()
#ask if the user want to calculate again
def again():
    calc_again = input('''
Do you want to calculate again?
Please type Y for YES or exit for exiting the calculator:
''')

    if calc_again.upper() == 'Y':
        calculate()
    elif calc_again.lower() == 'exit':
        print('Exiting calculator. See you later, Goodbye')#exit from the calculator
    else:
        again()

calculate()
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Promt 1. Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.


```python
import csv
import os

# Constants
INVENTORY_FILE = "inventory.csv"
CATEGORIES = ["Electronics", "Clothing", "Home & Kitchen", "Books", "Toys"]

# Ensure the CSV file exists and is not corrupted
def initialize_inventory():
    if not os.path.exists(INVENTORY_FILE):
        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["ProductID", "Name", "Category", "Quantity", "Price"])
    else:
        try:
            with open(INVENTORY_FILE, mode="r") as file:
                reader = csv.reader(file)
                try:
                    headers = next(reader)
                    if headers != ["ProductID", "Name", "Category", "Quantity", "Price"]:
                        raise ValueError("Corrupted CSV file: Invalid headers.")
                except StopIteration:  # File is empty
                    raise ValueError("Corrupted CSV file: File is empty.")
        except Exception as e:
            print(f"Error: {e}. Reinitializing inventory file.")
            with open(INVENTORY_FILE, mode="w", newline="") as file:
                writer = csv.writer(file)
                writer.writerow(["ProductID", "Name", "Category", "Quantity", "Price"])

# Add a new product to the inventory
def add_product(product_id, name, category, quantity, price):
    # Check for duplicate Product ID
    with open(INVENTORY_FILE, mode="r") as file:
        reader = csv.reader(file)
        next(reader)  # Skip headers
        for row in reader:
            if row[0] == product_id:
                print(f"Error: Product ID '{product_id}' already exists.")
                return

    # Validate category
    if category not in CATEGORIES:
        print(f"Error: Invalid category. Must be one of {CATEGORIES}.")
        return

    # Validate quantity and price
    try:
        quantity = int(quantity)
        price = float(price)
        if quantity < 0 or price < 0:
            print("Error: Quantity and price must be non-negative.")
            return
    except ValueError:
        print("Error: Quantity must be an integer and price must be a number.")
        return

    # Add the product
    try:
        with open(INVENTORY_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([product_id, name, category, quantity, price])
        print(f"Product '{name}' added successfully.")
    except PermissionError:
        print("Error: Permission denied. Cannot write to the file.")

# Delete a product from the inventory
def delete_product(product_id):
    temp_file = INVENTORY_FILE + ".tmp"
    found = False

    try:
        with open(INVENTORY_FILE, mode="r") as infile, open(temp_file, mode="w", newline="") as outfile:
            reader = csv.reader(infile)
            writer = csv.writer(outfile)
            headers = next(reader)
            writer.writerow(headers)
            for row in reader:
                if row[0] == product_id:
                    found = True
                else:
                    writer.writerow(row)

        if not found:
            print(f"Error: Product with ID '{product_id}' not found.")
            os.remove(temp_file)
            return

        os.replace(temp_file, INVENTORY_FILE)
        print(f"Product with ID '{product_id}' deleted successfully.")
    except Exception as e:
        print(f"Error: {e}. Failed to delete product.")

# Update product details
def update_product(product_id, name=None, category=None, quantity=None, price=None):
    rows = []
    found = False

    try:
        with open(INVENTORY_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == product_id:
                    found = True
                    if name:
                        row[1] = name
                    if category:
                        if category not in CATEGORIES:
                            print(f"Error: Invalid category. Must be one of {CATEGORIES}.")
                            return
                        row[2] = category
                    if quantity:
                        try:
                            quantity = int(quantity)
                            if quantity < 0:
                                print("Error: Quantity must be non-negative.")
                                return
                            row[3] = quantity
                        except ValueError:
                            print("Error: Quantity must be an integer.")
                            return
                    if price:
                        try:
                            price = float(price)
                            if price < 0:
                                print("Error: Price must be non-negative.")
                                return
                            row[4] = price
                        except ValueError:
                            print("Error: Price must be a number.")
                            return
                rows.append(row)

        if not found:
            print(f"Error: Product with ID '{product_id}' not found.")
            return

        with open(INVENTORY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)
        print(f"Product with ID '{product_id}' updated successfully.")
    except Exception as e:
        print(f"Error: {e}. Failed to update product.")

# Search for products by name or category
def search_products(search_term, by_category=False):
    results = []

    try:
        with open(INVENTORY_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                if by_category:
                    if row[2].lower() == search_term.lower():
                        results.append(row)
                else:
                    if search_term.lower() in row[1].lower():
                        results.append(row)

        if not results:
            print("No products found.")
        else:
            print("Search Results:")
            for result in results:
                print(result)
    except Exception as e:
        print(f"Error: {e}. Failed to search products.")

# Generate a report showing item values and category breakdowns
def generate_report():
    category_totals = {category: {"count": 0, "value": 0.0} for category in CATEGORIES}
    total_value = 0.0

    try:
        with open(INVENTORY_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                if len(row) != 5:  # Skip incomplete rows
                    continue
                category = row[2]
                quantity = int(row[3])
                price = float(row[4])
                value = quantity * price

                if category in category_totals:
                    category_totals[category]["count"] += quantity
                    category_totals[category]["value"] += value
                    total_value += value

        print("\nInventory Report:")
        print(f"{'Category':<15} {'Quantity':<10} {'Total Value':<15}")
        for category, data in category_totals.items():
            print(f"{category:<15} {data['count']:<10} ${data['value']:.2f}")
        print(f"\nTotal Inventory Value: ${total_value:.2f}")
    except Exception as e:
        print(f"Error: {e}. Failed to generate report.")

# Main menu
def main_menu():
    while True:
        print("\nInventory Management System")
        print("1. Add Product")
        print("2. Delete Product")
        print("3. Update Product")
        print("4. Search Products")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            product_id = input("Enter Product ID: ")
            name = input("Enter Product Name: ")
            category = input(f"Enter Category ({', '.join(CATEGORIES)}): ")
            quantity = input("Enter Quantity: ")
            price = input("Enter Price: ")
            add_product(product_id, name, category, quantity, price)

        elif choice == "2":
            product_id = input("Enter Product ID to delete: ")
            delete_product(product_id)

        elif choice == "3":
            product_id = input("Enter Product ID to update: ")
            name = input("Enter new Name (leave blank to skip): ")
            category = input(f"Enter new Category ({', '.join(CATEGORIES)}) (leave blank to skip): ")
            quantity = input("Enter new Quantity (leave blank to skip): ")
            price = input("Enter new Price (leave blank to skip): ")
            update_product(product_id, name or None, category or None, quantity or None, price or None)

        elif choice == "4":
            search_type = input("Search by (1) Name or (2) Category? ")
            if search_type == "1":
                search_term = input("Enter search term: ")
                search_products(search_term)
            elif search_type == "2":
                search_term = input(f"Enter category ({', '.join(CATEGORIES)}): ")
                search_products(search_term, by_category=True)
            else:
                print("Invalid choice.")

        elif choice == "5":
            generate_report()

        elif choice == "6":
            print("Exiting the system. See you Later, Goodbye")
            break

        else:
            print("Invalid choice. Please try again.")

# Initialize and run the program
if __name__ == "__main__":
    initialize_inventory()
    main_menu()
```

```python
import csv
import os
from datetime import datetime, timedelta

# Constants
FLIGHTS_FILE = "flights.csv"
BOOKINGS_FILE = "bookings.csv"
LOYALTY_FILE = "loyalty.csv"
SEAT_TYPES = ["Window", "Aisle", "Middle"]
FLIGHT_STATUSES = ["On-time", "Delayed", "Cancelled"]

# Ensure the CSV files exist and are not corrupted
def initialize_files():
    # Initialize flights file
    if not os.path.exists(FLIGHTS_FILE):
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable", "Status"])
        generate_sample_flights()  # Generate sample flight data

    # Initialize bookings file
    if not os.path.exists(BOOKINGS_FILE):
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["BookingID", "FlightID", "PassengerName", "SeatNumber", "SeatType", "BookingTime"])

    # Initialize loyalty file
    if not os.path.exists(LOYALTY_FILE):
        with open(LOYALTY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["PassengerName", "LoyaltyPoints"])
        generate_sample_loyalty()  # Generate sample loyalty data

# Generate sample flight data
def generate_sample_flights():
    sample_flights = [
        ["F101", "New York", "Los Angeles", (datetime.now() + timedelta(days=1)).strftime("%Y-%m-%d %H:%M"), 100, "On-time"],
        ["F102", "Chicago", "Miami", (datetime.now() + timedelta(days=2)).strftime("%Y-%m-%d %H:%M"), 150, "On-time"],
        ["F103", "San Francisco", "Seattle", (datetime.now() + timedelta(days=3)).strftime("%Y-%m-%d %H:%M"), 200, "Delayed"],
        ["F104", "Boston", "Atlanta", (datetime.now() + timedelta(days=4)).strftime("%Y-%m-%d %H:%M"), 120, "On-time"],
        ["F105", "Dallas", "Denver", (datetime.now() + timedelta(days=5)).strftime("%Y-%m-%d %H:%M"), 180, "On-time"],
    ]
    with open(FLIGHTS_FILE, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(sample_flights)
    print("Sample flight data generated.")

# Generate sample loyalty data
def generate_sample_loyalty():
    sample_loyalty = [
        ["John Doe", 50],
        ["Jane Smith", 100],
        ["Alice Johnson", 75],
    ]
    with open(LOYALTY_FILE, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(sample_loyalty)
    print("Sample loyalty data generated.")

# Add a new flight
def add_flight(flight_id, origin, destination, departure_time, seats_available, status="On-time"):
    # Check for duplicate Flight ID
    with open(FLIGHTS_FILE, mode="r") as file:
        reader = csv.reader(file)
        next(reader)  # Skip headers
        for row in reader:
            if row[0] == flight_id:
                print(f"Error: Flight ID '{flight_id}' already exists.")
                return

    # Add the flight
    try:
        with open(FLIGHTS_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([flight_id, origin, destination, departure_time, seats_available, status])
        print(f"Flight '{flight_id}' added successfully.")
    except Exception as e:
        print(f"Error: {e}. Failed to add flight.")

# Find available flights
def find_flights(origin, destination):
    available_flights = []

    try:
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                if row[1].lower() == origin.lower() and row[2].lower() == destination.lower() and int(row[4]) > 0:
                    available_flights.append(row)

        if not available_flights:
            print("No available flights found.")
        else:
            print("Available Flights:")
            for flight in available_flights:
                print(flight)
    except Exception as e:
        print(f"Error: {e}. Failed to find flights.")

# Reserve a seat
def reserve_seat(flight_id, passenger_name, seat_type):
    # Validate seat type
    if seat_type not in SEAT_TYPES:
        print(f"Error: Invalid seat type. Must be one of {SEAT_TYPES}.")
        return

    # Check flight availability
    rows = []
    found = False
    try:
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == flight_id:
                    found = True
                    seats_available = int(row[4])
                    if seats_available <= 0:
                        print("Error: No seats available on this flight.")
                        return
                    row[4] = str(seats_available - 1)  # Reduce available seats
                rows.append(row)

        if not found:
            print(f"Error: Flight ID '{flight_id}' not found.")
            return

        # Update flights file
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)

        # Add booking
        booking_id = datetime.now().strftime("%Y%m%d%H%M%S")
        booking_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open(BOOKINGS_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([booking_id, flight_id, passenger_name, "", seat_type, booking_time])

        # Update loyalty points
        update_loyalty_points(passenger_name, 10)  # Award 10 points per booking

        print(f"Seat reserved successfully. Booking ID: {booking_id}")
    except Exception as e:
        print(f"Error: {e}. Failed to reserve seat.")

# Release a seat
def release_seat(booking_id):
    rows = []
    found = False
    try:
        with open(BOOKINGS_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == booking_id:
                    found = True
                    flight_id = row[1]
                else:
                    rows.append(row)

        if not found:
            print(f"Error: Booking ID '{booking_id}' not found.")
            return

        # Update bookings file
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)

        # Increase available seats in flights file
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            flight_rows = list(reader)

        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(flight_rows[0])  # Write headers
            for row in flight_rows[1:]:
                if row[0] == flight_id:
                    row[4] = str(int(row[4]) + 1)  # Increase available seats
                writer.writerow(row)

        print(f"Seat released successfully for Booking ID: {booking_id}")
    except Exception as e:
        print(f"Error: {e}. Failed to release seat.")

# Update loyalty points
def update_loyalty_points(passenger_name, points):
    rows = []
    found = False
    try:
        with open(LOYALTY_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == passenger_name:
                    found = True
                    row[1] = str(int(row[1]) + points)
                rows.append(row)

        if not found:
            rows.append([passenger_name, str(points)])

        # Update loyalty file
        with open(LOYALTY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)
    except Exception as e:
        print(f"Error: {e}. Failed to update loyalty points.")

# Generate performance report
def generate_report():
    total_bookings = 0
    total_revenue = 0
    flight_bookings = {}

    try:
        with open(BOOKINGS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                total_bookings += 1
                flight_id = row[1]
                if flight_id in flight_bookings:
                    flight_bookings[flight_id] += 1
                else:
                    flight_bookings[flight_id] = 1

        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                flight_id = row[0]
                seats_available = int(row[4])
                total_seats = seats_available + flight_bookings.get(flight_id, 0)
                total_revenue += flight_bookings.get(flight_id, 0) * 100  # Assume $100 per ticket

        print("\nPerformance Report:")
        print(f"Total Bookings: {total_bookings}")
        print(f"Total Revenue: ${total_revenue}")
        print("\nBookings per Flight:")
        for flight_id, bookings in flight_bookings.items():
            print(f"Flight {flight_id}: {bookings} bookings")
    except Exception as e:
        print(f"Error: {e}. Failed to generate report.")

# Main menu
def main_menu():
    while True:
        print("\nFlight Booking System")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Reserve Seat")
        print("4. Release Seat")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            flight_id = input("Enter Flight ID: ")
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            departure_time = input("Enter Departure Time (YYYY-MM-DD HH:MM): ")
            seats_available = input("Enter Seats Available: ")
            add_flight(flight_id, origin, destination, departure_time, seats_available)

        elif choice == "2":
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            find_flights(origin, destination)

        elif choice == "3":
            flight_id = input("Enter Flight ID: ")
            passenger_name = input("Enter Passenger Name: ")
            seat_type = input(f"Enter Seat Type ({', '.join(SEAT_TYPES)}): ")
            reserve_seat(flight_id, passenger_name, seat_type)

        elif choice == "4":
            booking_id = input("Enter Booking ID: ")
            release_seat(booking_id)

        elif choice == "5":
            generate_report()

        elif choice == "6":
            print("Exiting the system.")
            break

        else:
            print("Invalid choice. Please try again.")

# Initialize and run the program
if __name__ == "__main__":
    initialize_files()
    main_menu()
```

```python
import csv
import os
from datetime import datetime, timedelta

# Constants
FLIGHTS_FILE = "flights.csv"
BOOKINGS_FILE = "bookings.csv"
LOYALTY_FILE = "loyalty.csv"
SEAT_TYPES = ["Window", "Aisle", "Middle"]
FLIGHT_STATUSES = ["On-time", "Delayed", "Cancelled"]

# Ensure the CSV files exist and are not corrupted
def initialize_files():
    # Initialize flights file
    if not os.path.exists(FLIGHTS_FILE):
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["FlightID", "Origin", "Destination", "DepartureTime", "SeatsAvailable", "Status"])
        generate_sample_flights()  # Generate sample flight data

    # Initialize bookings file
    if not os.path.exists(BOOKINGS_FILE):
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["BookingID", "FlightID", "PassengerName", "SeatNumber", "SeatType", "BookingTime"])

    # Initialize loyalty file
    if not os.path.exists(LOYALTY_FILE):
        with open(LOYALTY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(["PassengerName", "LoyaltyPoints"])
        generate_sample_loyalty()  # Generate sample loyalty data

# Generate sample flight data
def generate_sample_flights():
    sample_flights = [
        ["F101", "New York", "Los Angeles", (datetime.now() + timedelta(days=1)).strftime("%Y-%m-%d %H:%M"), 100, "On-time"],
        ["F102", "Chicago", "Miami", (datetime.now() + timedelta(days=2)).strftime("%Y-%m-%d %H:%M"), 150, "On-time"],
        ["F103", "San Francisco", "Seattle", (datetime.now() + timedelta(days=3)).strftime("%Y-%m-%d %H:%M"), 200, "Delayed"],
        ["F104", "Boston", "Atlanta", (datetime.now() + timedelta(days=4)).strftime("%Y-%m-%d %H:%M"), 120, "On-time"],
        ["F105", "Dallas", "Denver", (datetime.now() + timedelta(days=5)).strftime("%Y-%m-%d %H:%M"), 180, "On-time"],
    ]
    with open(FLIGHTS_FILE, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(sample_flights)
    print("Sample flight data generated.")

# Generate sample loyalty data
def generate_sample_loyalty():
    sample_loyalty = [
        ["John Doe", 50],
        ["Jane Smith", 100],
        ["Alice Johnson", 75],
    ]
    with open(LOYALTY_FILE, mode="a", newline="") as file:
        writer = csv.writer(file)
        writer.writerows(sample_loyalty)
    print("Sample loyalty data generated.")

# Add a new flight
def add_flight(flight_id, origin, destination, departure_time, seats_available, status="On-time"):
    # Check for duplicate Flight ID
    with open(FLIGHTS_FILE, mode="r") as file:
        reader = csv.reader(file)
        next(reader)  # Skip headers
        for row in reader:
            if row[0] == flight_id:
                print(f"Error: Flight ID '{flight_id}' already exists.")
                return

    # Add the flight
    try:
        with open(FLIGHTS_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([flight_id, origin, destination, departure_time, seats_available, status])
        print(f"Flight '{flight_id}' added successfully.")
    except Exception as e:
        print(f"Error: {e}. Failed to add flight.")

# Find available flights
def find_flights(origin, destination):
    available_flights = []

    try:
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                if row[1].lower() == origin.lower() and row[2].lower() == destination.lower() and int(row[4]) > 0:
                    available_flights.append(row)

        if not available_flights:
            print("No available flights found.")
        else:
            print("Available Flights:")
            for flight in available_flights:
                print(flight)
    except Exception as e:
        print(f"Error: {e}. Failed to find flights.")

# Reserve a seat
def reserve_seat(flight_id, passenger_name, seat_type):
    # Validate seat type
    if seat_type not in SEAT_TYPES:
        print(f"Error: Invalid seat type. Must be one of {SEAT_TYPES}.")
        return

    # Check flight availability
    rows = []
    found = False
    try:
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == flight_id:
                    found = True
                    seats_available = int(row[4])
                    if seats_available <= 0:
                        print("Error: No seats available on this flight.")
                        return
                    row[4] = str(seats_available - 1)  # Reduce available seats
                rows.append(row)

        if not found:
            print(f"Error: Flight ID '{flight_id}' not found.")
            return

        # Update flights file
        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)

        # Add booking
        booking_id = datetime.now().strftime("%Y%m%d%H%M%S")
        booking_time = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        with open(BOOKINGS_FILE, mode="a", newline="") as file:
            writer = csv.writer(file)
            writer.writerow([booking_id, flight_id, passenger_name, "", seat_type, booking_time])

        # Update loyalty points
        update_loyalty_points(passenger_name, 10)  # Award 10 points per booking

        print(f"Seat reserved successfully. Booking ID: {booking_id}")
    except Exception as e:
        print(f"Error: {e}. Failed to reserve seat.")

# Release a seat
def release_seat(booking_id):
    rows = []
    found = False
    try:
        with open(BOOKINGS_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == booking_id:
                    found = True
                    flight_id = row[1]
                else:
                    rows.append(row)

        if not found:
            print(f"Error: Booking ID '{booking_id}' not found.")
            return

        # Update bookings file
        with open(BOOKINGS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)

        # Increase available seats in flights file
        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            flight_rows = list(reader)

        with open(FLIGHTS_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(flight_rows[0])  # Write headers
            for row in flight_rows[1:]:
                if row[0] == flight_id:
                    row[4] = str(int(row[4]) + 1)  # Increase available seats
                writer.writerow(row)

        print(f"Seat released successfully for Booking ID: {booking_id}")
    except Exception as e:
        print(f"Error: {e}. Failed to release seat.")

# Update loyalty points
def update_loyalty_points(passenger_name, points):
    rows = []
    found = False
    try:
        with open(LOYALTY_FILE, mode="r") as file:
            reader = csv.reader(file)
            headers = next(reader)
            for row in reader:
                if row[0] == passenger_name:
                    found = True
                    row[1] = str(int(row[1]) + points)
                rows.append(row)

        if not found:
            rows.append([passenger_name, str(points)])

        # Update loyalty file
        with open(LOYALTY_FILE, mode="w", newline="") as file:
            writer = csv.writer(file)
            writer.writerow(headers)
            writer.writerows(rows)
    except Exception as e:
        print(f"Error: {e}. Failed to update loyalty points.")

# Generate performance report
def generate_report():
    total_bookings = 0
    total_revenue = 0
    flight_bookings = {}

    try:
        with open(BOOKINGS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                total_bookings += 1
                flight_id = row[1]
                if flight_id in flight_bookings:
                    flight_bookings[flight_id] += 1
                else:
                    flight_bookings[flight_id] = 1

        with open(FLIGHTS_FILE, mode="r") as file:
            reader = csv.reader(file)
            next(reader)  # Skip headers
            for row in reader:
                flight_id = row[0]
                seats_available = int(row[4])
                total_seats = seats_available + flight_bookings.get(flight_id, 0)
                total_revenue += flight_bookings.get(flight_id, 0) * 100  # Assume $100 per ticket

        print("\nPerformance Report:")
        print(f"Total Bookings: {total_bookings}")
        print(f"Total Revenue: ${total_revenue}")
        print("\nBookings per Flight:")
        for flight_id, bookings in flight_bookings.items():
            print(f"Flight {flight_id}: {bookings} bookings")
    except Exception as e:
        print(f"Error: {e}. Failed to generate report.")

# Main menu
def main_menu():
    while True:
        print("\nFlight Booking System")
        print("1. Add Flight")
        print("2. Find Flights")
        print("3. Reserve Seat")
        print("4. Release Seat")
        print("5. Generate Report")
        print("6. Exit")
        choice = input("Enter your choice: ")

        if choice == "1":
            flight_id = input("Enter Flight ID: ")
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            departure_time = input("Enter Departure Time (YYYY-MM-DD HH:MM): ")
            seats_available = input("Enter Seats Available: ")
            add_flight(flight_id, origin, destination, departure_time, seats_available)

        elif choice == "2":
            origin = input("Enter Origin: ")
            destination = input("Enter Destination: ")
            find_flights(origin, destination)

        elif choice == "3":
            flight_id = input("Enter Flight ID: ")
            passenger_name = input("Enter Passenger Name: ")
            seat_type = input(f"Enter Seat Type ({', '.join(SEAT_TYPES)}): ")
            reserve_seat(flight_id, passenger_name, seat_type)

        elif choice == "4":
            booking_id = input("Enter Booking ID: ")
            release_seat(booking_id)

        elif choice == "5":
            generate_report()

        elif choice == "6":
            print("Exiting the system.")
            break

        else:
            print("Invalid choice. Please try again.")

# Initialize and run the program
if __name__ == "__main__":
    initialize_files()
    main_menu()
```
